```python
import matplotlib.pyplot as plt

import pandas as pd 
import numpy as np
```

```python
x=[45,25,44,35]
y=["Java","Html","css","BootStrap"]


plt.plot(y,x,marker="*",color="y",markersize=20,linewidth=2,linestyle="-.",markeredgecolor="r")

plt.grid()
plt.show()
```

```python
x=[45,25,44,35]
y=["Java","Html","css","BootStrap"]
plt.barh(y,x,color=["r","g","y","k"],align="edge",alpha=0.2)
plt.show()
```

```python
x=[45,25,44,35]
y=["Java","Html","css","BootStrap"]
plt.bar(y,x,color=["r","g","y","k"],align="edge",alpha=0.2,width=0.5,edgecolor="k",linewidth=2,linestyle=":")
plt.title("This is Bar Chart")
plt.show()
```
