```python
x=["Html","Java","Python","CSS"]
y=[45,78,65,54]
c=["y","r","b","g"]
plt.bar(x,y,width=0.4,color=c,align='edge')
plt.title("Bar Graph",fontsize=30,color="red")
plt.xlabel("Courses",fontsize=20,color="blue")
# plt.legend("Course Bar")
plt.ylabel("Number")
# plt.savefig("BarGraph.png",dpi=200)
plt.show()
```

```python
x=["Html","Java","Python","CSS"]
y=[45,78,65,54]
c=["y","r","b","g"]
plt.bar(x,y,width=0.4,color=c,align='edge',edgecolor="y",linewidth=5,alpha=0.2)  

# width for bar width
# align= for lebel 
# edgecolor for bar border color
# linewidth for bar color px 
# alpha for bar blueness

plt.title("Bar Graph",fontsize=30,color="red")
plt.xlabel("Courses",fontsize=20,color="blue")
plt.ylabel("Number")
plt.show()
```

```python
a=[20,30,40,50,10]
plt.boxplot(a,vert=False,widths=0.3,labels=["Python"],patch_artist=True)
plt.title("This is BoxPlot",color="r",fontsize=30,alpha=0.4,style="italic")
# plt.legend("Python" ,loc=3)
# plt.savefig("Boxplot",dpi=200,facecolor="y",transparent=True)
plt.show()
```

```python
a=[20,30,40,50,10]
plt.boxplot(a,vert=False,widths=0.3,labels=["Python"],showmeans=True)
plt.title("This is BoxPlot",color="r",fontsize=30,alpha=0.4,style="italic")
plt.show()
```

```python
x=[1,2,3,4,5]
q=[1,2,3,4,5]
w=[1,2,3,4,5]
e=[2,1,3,4,5]
l=["Area1","Area2","Area3"]
plt.stackplot(x,q,w,e,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
a=[10,20,30,40,50]
y=[20,30,40,50,60]
plt.step(a,y,marker="*",markeredgecolor="r",markersize=20,linestyle=":",label=["Python"])
plt.title("This is Step Chart",color="y",fontsize=20,style="italic")
plt.legend(loc=2)
plt.grid()
plt.savefig("Step1",dpi=200,facecolor="r",transparent=True)
plt.show()
```
