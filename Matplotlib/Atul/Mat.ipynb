```python
plt.figure(figsize=(7,5))

a=[10,11,30,40]
y=["Html","Java","Css","Jascripts"]
plt.plot(y,a,marker="*",color="r",markersize=20,mfc="y",linewidth=2,linestyle="-.",label=["Python Number"])
plt.legend()
plt.title("This is Python Module Number",color="y",style="italic",fontsize=20)
plt.xlabel("Subject",color="r",fontsize=15)
plt.ylabel("Number",color="r",fontsize=15)
plt.grid()
plt.savefig("Python",facecolor="y",dpi=200,transparent=True)
plt.show()
```

```python
plt.figure(figsize=(7,5))

Subject=["Html","Java","Css","Javascripts","Mysql","Jquery"]
marks=[40,45,35,30,25,28]
color=["r","b","g","y","k","c"]

plt.bar(Subject,marks,color=color,edgecolor="r",linewidth=5,alpha=0.4)

plt.xlabel("Subject",color="r",fontsize=15)
plt.ylabel("Number",color="r",fontsize=15)

plt.savefig("Python Bar Chart.pdf",dpi=200,facecolor="g")

plt.show()
```

```python
plt.figure(figsize=(7,5))

Subject=["Html","Java","Css","Jascripts","Mysql","Jquery"]
marks=[40,45,35,30,25,28]
color=["r","b","g","y","k","c"]

plt.barh(Subject,marks,color=color,edgecolor="r",linewidth=5,alpha=0.4)

plt.show()
```
